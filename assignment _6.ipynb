```python
# TIC-TAC-TOE GAME (2 Player Console Version)

# Function to print the board
def print_board(board):
    print("\n")
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print("\n")

# Function to check if a player has won
def check_winner(board, player):
    # All possible winning combinations (rows, columns, diagonals)
    win_conditions = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
        (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
        (0, 4, 8), (2, 4, 6)              # Diagonals
    ]

    # Check each combination
    for a, b, c in win_conditions:
        if board[a] == board[b] == board[c] == player:
            return True
    return False

# Function to check for a draw (no empty spaces)
def check_draw(board):
    return all(cell != " " for cell in board)

# Main function for the game
def tic_tac_toe():
    board = [" "] * 9        # Create empty board with 9 cells
    current_player = "X"     # Player X starts first

    print("Welcome to Tic-Tac-Toe Game!")
    print_board(board)

    while True:
        try:
            # Take input from current player
            move = int(input(f"Player {current_player}, enter your move (1-9): ")) - 1

            # Validate move (must be inside the board and empty)
            if move < 0 or move > 8 or board[move] != " ":
                print("Invalid move! Try again.")
                continue

            # Place the player's symbol on the board
            board[move] = current_player
            print_board(board)

            # Check for winner
            if check_winner(board, current_player):
                print(f"Player {current_player} wins!")
                break

            # Check for draw
            if check_draw(board):
                print("It's a Draw!")
                break

            # Switch player after each valid move
            current_player = "O" if current_player == "X" else "X"

        except ValueError:
            # If player enters something that is not a number
            print("Please enter a valid number between 1 and 9.")

# Run the game
tic_tac_toe()
```
